# **DATA SPLITTING**

```python
#Input berupa seluruh data selain kolom NObeyesdad, kolom NObeyesdad sebagai kolom output / target
input = df.drop('NObeyesdad', axis=1)
output = df['NObeyesdad']
```

```python
#Split data menjadi train (80%) dan test (20%)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input, output, test_size = 0.2, random_state = 42)
```

# **DATA CLEANING**

```python
#Hapus ' years', ubah tipe data ke integer
x_train['Age'] = x_train['Age'].str.replace(' years', '', regex=False).astype(int)
x_test['Age'] = x_test['Age'].str.replace(' years', '', regex=False).astype(int)
```

```python
#Cek outlier
boxplot = x_train.boxplot(column=['Height', 'FCVC',	'NCP', 'CH2O', 'FAF', 'TUE'])
plt.show()
```

```python
#Cek outlier
boxplot = x_train.boxplot(column=['Age'])
plt.show()
```

```python
#Cek outlier
boxplot = x_train.boxplot(column=['Weight'])
plt.show()
```

```python
#Hitung mean variabel FCVC
x_train['FCVC'].mean()
```

```python
#Ubah missing value di variable FCVC menjadi nilai meannya
x_train['FCVC'] = x_train['FCVC'].fillna(x_train['FCVC'].mean())
x_test['FCVC'] = x_test['FCVC'].fillna(x_train['FCVC'].mean())
```

```python
#Cari modus variabel MTRANS
x_train['MTRANS'].mode()
```

```python
#Ubah missing value di variabel MTRANS menjadi nilai modusnya
x_train['MTRANS'] = x_train['MTRANS'].fillna(x_train['MTRANS'].mode()[0])
x_test['MTRANS'] = x_test['MTRANS'].fillna(x_train['MTRANS'].mode()[0])
```

# **ENCODING**

```python
#Binary Encoding -> hanya ada 2 output
binary_encode = {"Gender": {"Male":1,"Female" :0}, "family_history_with_overweight":{"yes":1,"no":0}, "FAVC":{"yes":1,"no":0}, "SMOKE":{"yes":1,"no":0}, "SCC":{"yes":1,"no":0}}
x_train = x_train.replace(binary_encode)
x_test = x_test.replace(binary_encode)
```

```python
x_train
```

```python
#Ordinal Encoding -> ada urutannya
mapping = {
    'no': 0,
    'Sometimes': 1,
    'Frequently': 2,
    'Always': 3
}

x_train['CAEC'] = x_train['CAEC'].replace(mapping)
x_test['CAEC'] = x_test['CAEC'].replace(mapping)
x_train['CALC'] = x_train['CALC'].replace(mapping)
x_test['CALC'] = x_test['CALC'].replace(mapping)
```

```python
x_train
```

```python
#One-Hot Encoding -> ga ada urutannya & lebih dari dua kelas
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
xtrain_enc = pd.DataFrame(enc.fit_transform(x_train[['MTRANS']]).toarray(), columns=enc.get_feature_names_out(), index=x_train.index)
xtest_enc = pd.DataFrame(enc.transform(x_test[['MTRANS']]).toarray(), columns=enc.get_feature_names_out(), index=x_test.index)

x_train = pd.concat([x_train.drop('MTRANS', axis=1), xtrain_enc], axis=1)
x_test = pd.concat([x_test.drop('MTRANS', axis=1), xtest_enc], axis=1)
```

```python
#Label Encoding -> lebih dari 2 kelas, secara ga langsung ada urutannya
from sklearn import preprocessing
label_encoding = preprocessing.LabelEncoder()
y_train=label_encoding.fit_transform(y_train)
y_test=label_encoding.transform(y_test)
```

# **RANDOM FOREST**

```python
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')

RF_class = RandomForestClassifier(criterion= 'gini', max_depth = 10)
RF_class.fit(x_train, y_train)
```

```python
#Cek akurasi train
RF_class.score(x_train, y_train)
```

```python
#Cek akurasi test
RF_class.score(x_test, y_test)
```

```python
#Prediksi output berdasarkan data test
y_pred_rf = RF_class.predict(x_test)
```

```python
#Evaluasi
from sklearn.metrics import classification_report
print('\nClassification Report\n')
print(classification_report(y_test, y_pred_rf, target_names=['0','1','2','3','4','5','6']))
```

- Precision = seberapa tepat model memprediksi kelas tersebut.
- Recall = seberapa banyak total data kelas yang berhasil dikenali.
- F1-Score = menggambarkan keseimbangan antara precision dan recall
- Model mempunyai akurasi keseluruhan yang cukup tinggi yaitu 88%. Tapi, performa tersebut tidak merata yang ditunjukkan dengan nilai f1-score. Kelas 3 dan 4 mempunyai f1-score yang mendekati 1.00 menunjukkan model akurat dan jarang salah mendeteksi. Sedangkan kelas 1 dan 6 mempunyai f1-score rendah yaitu 0.77 menunjukkan model masih belum bisa memprediksi dengan baik. Secara keseluruhan, model sudah cukup baik namun perlu perbaikan untuk kelas 1 dan 6, mungkin bisa dengan menambah data, menganalisis fitur yang membedakan kelas tersebut, ataupun lainnya.

# **XGBoost**

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')

xgb_model = XGBClassifier(max_depth=15, n_estimators=200, use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(x_train, y_train)
```

```python
#Cek akurasi train
xgb_model.score(x_train, y_train)
```

```python
#Cek akurasi test
xgb_model.score(x_test, y_test)
```

```python
#Prediksi output berdasarkan data test
y_pred_xgb = xgb_model.predict(x_test)
```

```python
#Evaluasi
from sklearn.metrics import classification_report
print('\nClassification Report\n')
print(classification_report(y_test, y_pred_xgb, target_names=['0','1','2','3','4','5','6']))
```

- Precision = seberapa tepat model memprediksi kelas tersebut.
- Recall = seberapa banyak total data kelas yang berhasil dikenali.
- F1-Score = menggambarkan keseimbangan antara precision dan recall
- Model mempunyai akurasi keseluruhan yang cukup tinggi yaitu 95% yang berarti model dapat mengklasifikasikan 95% dari total 212 data test dengan benar. Tidak ada kelas yang memiliki precision atau recall yang sangat buruk (semuanya diatas 0.88) menunjukkan model cukup stabil. Kelas terbaik yaitu kelas 4 dengan precision, recall, dan f1-score sempurna yaitu 1.00 yang berarti tidak ada kesalahan klasifikasi untuk kelas 4. Namun ada kelas yang performanya cukup kurang secara relatif yaitu kelas 5 dengan f1-score 0.90 dan recall 0.88. Secara keseluruhan, model sudah cukup baik dan seimbang dalam memprediksi seluruh kelas.
